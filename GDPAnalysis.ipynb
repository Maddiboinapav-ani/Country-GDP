# Introduction to seaborn # seaborn is very powerfull visualizatio(STATISTIC VISULAIZATION) pkg in python


```python
import matplotlib.pyplot as plt # visulaiztion
import seaborn as sns # distribution visualtion 

%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4

import warnings
warnings.filterwarnings('ignore')
```

```python
states.head()
```

```python
# Distributions: 
# this combines histogram and kde
# Bin Width: The x-axis is divided into bins of equal width.
# Height: The height of each bar represents the number of data points that fall within the corresponding bin.
# Frequency: The taller the bar, the more data points are in that bin.
vis1 = sns.distplot(states["InternetUsers"]) 
```

```python
vis1 = sns.distplot(states["InternetUsers"], bins=10) # bins =10 says plot with 10 boxes
```

```python
#BOX PLOTS:
vis2 = sns.boxplot(data = states, x="IncomeGroup", y='BirthRate') # that circle denotes outliers and middle line indicates the avg income
```

```python
# visualizing with seaborn
```

```python
vis3 = sns.lmplot(data = states,x = 'InternetUsers', y = 'BirthRate', fit_reg = False) #lm - linear model fit_reg by default true here it is false indicating no linear relation ship
```

```python
vis4 = sns.lmplot(data = states,x = 'InternetUsers', y = 'BirthRate')
```

```python
vis5 = sns.lmplot(data = states,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = False,hue = 'IncomeGroup') #hue - parameter for color
```

```python
vis5 = sns.lmplot(data = states,x = 'InternetUsers', y = 'BirthRate',
                  fit_reg = False,hue = 'IncomeGroup', height = 4) # height parameter we can chage the image height
```
